```scala
import org.apache.spark.sql.{DataFrame, SparkSession}
import org.apache.spark.sql.functions._
import org.apache.spark.sql.types._

case class PressFE(
  enriched: DataFrame,      // DF enrichi avec nouvelles features
  rates: DataFrame,         // taux de retard par signe (-1,0,+1, null)
  ratesByBin: DataFrame     // taux de retard par (signe, bin de magnitude)
)

def buildPressFeatures(
  df: DataFrame,
  tendencyCol: String = "origin_weather_PressureTendency-1",
  changeCol: String   = "origin_weather_PressureChange-1",
  prevDelayCol: String = "PREV_AIRCRAFT_ARR_DELAY",
  visRiskCol: Option[String] = Some("origin_weather_feature_visibility_risk_score-1")
)(implicit spark: SparkSession): PressFE = {

  // 1) Signe de tendance: +1=hausse, 0=stable, -1=baisse, null=unknown
  val df1 = df
    .withColumn(
      "press_trend_sign",
      when(col(tendencyCol).isin("1","2","3"), lit( 1))
        .when(col(tendencyCol).isin("5","6","7","8"), lit(-1))
        .when(col(tendencyCol) === "0", lit(0))
        .otherwise(lit(null).cast(IntegerType))
    )
    .withColumn("press_change", col(changeCol).cast(DoubleType))

  // 2) Binning par quantiles (sur valeurs non nulles)
  val nonNullChange = df1.select(col("press_change")).na.drop()
  val quantiles: Array[Double] =
    if (nonNullChange.head(1).isEmpty) Array(0.0, 0.0, 0.0)
    else nonNullChange.stat.approxQuantile("press_change", Array(0.2, 0.5, 0.8), 1e-3)

  val q1 = quantiles(0); val q2 = quantiles(1); val q3 = quantiles(2)

  val df2 = df1.withColumn(
    "press_change_bin",
    when(col("press_change").isNull, lit("bin_na"))
      .when(col("press_change") <= q1, lit(s"$q1"))
      .when(col("press_change") <= q2, lit(s"($q1,$q2]"))
      .when(col("press_change") <= q3, lit(s"($q2,$q3]"))
      .otherwise(lit(s">$q3"))
  )

  // 3) Features numriques + interactions utiles
  val withNum = df2
    .withColumn("f_press_trend_sign", coalesce(col("press_trend_sign"), lit(0)))
    .withColumn("f_press_change_abs", coalesce(abs(col("press_change")), lit(0.0)))
    .withColumn(
      "f_prevDelay_x_falling",
      when(col("press_trend_sign") === -1, col(prevDelayCol)).otherwise(lit(0))
    )

  val enriched =
    visRiskCol.filter(c => withNum.columns.contains(c)).map { c =>
      withNum.withColumn(
        "f_visRisk_x_falling",
        when(col("press_trend_sign") === -1, col(c)).otherwise(lit(0))
      )
    }.getOrElse(withNum)

  // 4) Taux de retard par signe (lissage de Laplace)
  val base = enriched.groupBy("press_trend_sign", "is_delayed").count()

  val rates = base.groupBy("press_trend_sign")
    .pivot("is_delayed", Seq(0, 1)).sum("count")
    .na.fill(0, Seq("0","1"))
    .withColumn("total", col("0") + col("1"))
    .withColumn("delay_rate_pct", round( (col("1") + lit(1)) / (col("total") + lit(2)) * 100, 2))
    .orderBy(desc("delay_rate_pct"))

  // 5) Taux de retard par (signe, bin de magnitude)
  val byBin = enriched.groupBy("press_trend_sign", "press_change_bin", "is_delayed").count()

  val ratesByBin = byBin.groupBy("press_trend_sign", "press_change_bin")
    .pivot("is_delayed", Seq(0, 1)).sum("count")
    .na.fill(0, Seq("0","1"))
    .withColumn("total", col("0") + col("1"))
    .withColumn("delay_rate_pct", round( (col("1") + lit(1)) / (col("total") + lit(2)) * 100, 2))
    .orderBy(col("press_trend_sign").desc_nulls_last, desc("delay_rate_pct"))

  PressFE(enriched, rates, ratesByBin)
}
```

```scala
val fe = buildPressFeatures(
  df = joinedExplodedDF,  // ton DataFrame de vols
  tendencyCol = "origin_weather_PressureTendency-1",
  changeCol   = "origin_weather_PressureChange-1",
  prevDelayCol = "PREV_AIRCRAFT_ARR_DELAY",
  visRiskCol = Some("origin_weather_feature_visibility_risk_score-1") // ou None si absent
)

fe.rates.show(truncate = false)       // taux de retard par signe (-1,0,+1,null)
fe.ratesByBin.show(100, truncate=false) // par (signe, bin)
//fe.enriched.printSchema()             // vrifie les nouvelles colonnes
```

```scala
import org.apache.spark.sql.functions._

// ===== Cration des nouvelles features =====
val df_enriched = joinedExplodedDF
  // 1 Variation forte de pression (changement rapide de masse dair)
  .withColumn(
    "feature_pressure_variation_risk",
    when(abs(col("origin_weather_PressureChange-1")) > 20.0, 1).otherwise(0)
  )

  // 2 Catgorisation en niveaux de stabilit mto
  .withColumn(
    "feature_pressure_bucket",
    when(abs(col("origin_weather_PressureChange-1")) <= 9.0, lit("stable"))
      .when(abs(col("origin_weather_PressureChange-1")) <= 20.0, lit("transition"))
      .when(abs(col("origin_weather_PressureChange-1")) > 20.0, lit("perturbe"))
      .otherwise(lit("unknown"))
  )

  // 3 Interaction : pression x visibilit (si la colonne visibilit existe)
  .withColumn(
    "feature_pressure_visibility_combo",
    when(
      col("origin_weather_feature_visibility_risk_score-1").isNotNull,
      abs(col("origin_weather_PressureChange-1")) *
        col("origin_weather_feature_visibility_risk_score-1")
    ).otherwise(lit(0.0))
  )

df_enriched.select(
  "feature_pressure_variation_risk",
  "feature_pressure_bucket",
  "feature_pressure_visibility_combo"
).show(10, truncate = false)
```

```scala
val riskStats = df_enriched.groupBy("feature_pressure_variation_risk", "is_delayed")
  .count()
  .groupBy("feature_pressure_variation_risk")
  .pivot("is_delayed", Seq(0, 1))
  .sum("count")
  .na.fill(0)
  .withColumn("total", col("0") + col("1"))
  .withColumn("delay_rate_pct", round((col("1") + lit(1)) / (col("total") + lit(2)) * 100, 2))
  .orderBy(desc("delay_rate_pct"))

println("=== Taux de retard selon feature_pressure_variation_risk ===")
riskStats.show(truncate = false)
```

```scala
val bucketStats = df_enriched.groupBy("feature_pressure_bucket", "is_delayed")
  .count()
  .groupBy("feature_pressure_bucket")
  .pivot("is_delayed", Seq(0, 1))
  .sum("count")
  .na.fill(0)
  .withColumn("total", col("0") + col("1"))
  .withColumn("delay_rate_pct", round((col("1") + lit(1)) / (col("total") + lit(2)) * 100, 2))
  .orderBy(desc("delay_rate_pct"))

println("=== Taux de retard selon feature_pressure_bucket ===")
bucketStats.show(truncate = false)
```

```scala
// Calcul des bornes par quantiles
val qs = df_enriched.stat.approxQuantile("feature_pressure_visibility_combo", Array(0.25, 0.5, 0.75), 1e-3)
val (q1, q2, q3) = (qs(0), qs(1), qs(2))

val df_combo = df_enriched.withColumn(
  "feature_pressure_vis_combo_bin",
  when(col("feature_pressure_visibility_combo") <= q1, lit("faible"))
    .when(col("feature_pressure_visibility_combo") <= q2, lit("modere"))
    .when(col("feature_pressure_visibility_combo") <= q3, lit("eleve"))
    .otherwise(lit("tres_eleve"))
)

val comboStats = df_combo.groupBy("feature_pressure_vis_combo_bin", "is_delayed")
  .count()
  .groupBy("feature_pressure_vis_combo_bin")
  .pivot("is_delayed", Seq(0, 1))
  .sum("count")
  .na.fill(0)
  .withColumn("total", col("0") + col("1"))
  .withColumn("delay_rate_pct", round((col("1") + lit(1)) / (col("total") + lit(2)) * 100, 2))
  .orderBy(desc("delay_rate_pct"))

println("=== Taux de retard selon pression  visibilit ===")
comboStats.show(truncate = false)
```

```scala
val finalDf = df_enriched
  .withColumn("pressure_bucket_index",
    when(col("feature_pressure_bucket") === "stable", 0)
     .when(col("feature_pressure_bucket") === "transition", 1)
     .when(col("feature_pressure_bucket") === "perturbe", 2)
     .otherwise(3)
  )
```
