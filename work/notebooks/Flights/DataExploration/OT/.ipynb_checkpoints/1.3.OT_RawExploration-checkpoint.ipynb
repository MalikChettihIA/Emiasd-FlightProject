```scala
import org.apache.spark.sql.{DataFrame, SparkSession}
import org.apache.spark.sql.functions._
import org.apache.spark.sql.types._

import org.apache.spark.sql.{DataFrame, SparkSession}
import org.apache.spark.sql.functions._
import org.apache.spark.sql.types._

/**
 * CRITIQUE : Ajouter extends Serializable pour que Spark puisse sérialiser l'objet
 */
object WeatherTypeParser extends Serializable {
  
  case class WeatherCondition(
    intensity: String,
    hasThunderstorm: Int,
    hasFreezingConditions: Int,
    hasPrecipitation: Int,
    hasObscuration: Int,
    hasHazardousConditions: Int,
    specificPhenomena: Set[String]
  )
  
  // ============================================
  // DÉFINITION DES CODES MÉTÉO
  // ============================================
  
  val INTENSITY_CODES = Map(
    "+" -> "HEAVY",
    "-" -> "LIGHT",
    "" -> "MODERATE"
  )
  
  val DESCRIPTOR_CODES = Set(
    "MI", "BC", "PR", "DR", "BL", "SH", "FZ", "TS", "VC"
  )
  
  val PRECIPITATION_CODES = Set(
    "DZ", "RA", "SN", "GR", "GS", "PL", "SG", "IC", "UP"
  )
  
  val OBSCURATION_CODES = Set(
    "BR", "FG", "FU", "HZ", "DU", "SA", "VA", "PY"
  )
  
  val HAZARDOUS_CODES = Set(
    "DS", "SS", "FC", "SQ", "PO"
  )
  
  val FREEZING_PHENOMENA = Set(
    "FZRA", "FZDZ", "FZFG", "FZBR"
  )
  
  val COMPOSITE_CODES = Set(
    "TSRA", "TSSN", "TSDZ", "TSPL", "TSGR", "TSGS",
    "SHRA", "SHSN", "SHGR", "SHGS",
    "FZRA", "FZDZ", "FZFG", "FZBR",
    "BLSN", "BLDU", "BLSA",
    "DRDU", "DRSA", "DRSN",
    "BCFG", "MIFG", "PRFG",
    "VCFG", "VCSH", "VCTS",
    "RAGR", "RAGS", "RASN", "RAPL",
    "SNRA", "SNPL", "SNGS",
    "PLRA", "PLSN",
    "GSRA", "GRRA"
  )
  
  /**
   * Extrait tous les codes météo d'une chaîne WeatherType
   */
  def extractWeatherCodes(weatherType: String): Set[String] = {
    if (weatherType == null || weatherType.isEmpty) return Set.empty
    
    val cleaned = weatherType
      .replaceAll("^[+\\-]", "")
      .replaceAll("\\s+", " ")
      .trim
      .toUpperCase
    
    if (cleaned.isEmpty) return Set.empty
    
    val tokens = cleaned.split("\\s+")
    val allCodes = scala.collection.mutable.Set[String]()
    
    for (token <- tokens) {
      if (token.length >= 4) {
        COMPOSITE_CODES.find(token.startsWith) match {
          case Some(composite) =>
            allCodes += composite
            allCodes ++= extractComponentsFromComposite(composite)
            
            if (token.length > composite.length) {
              val remainder = token.substring(composite.length)
              allCodes ++= extractSimpleCodes(remainder)
            }
          case None =>
            allCodes ++= extractSimpleCodes(token)
        }
      } else {
        allCodes ++= extractSimpleCodes(token)
      }
    }
    
    allCodes.toSet
  }
  
  private def extractComponentsFromComposite(composite: String): Set[String] = {
    composite match {
      case s if s.startsWith("TS") => Set("TS", s.substring(2))
      case s if s.startsWith("SH") => Set("SH", s.substring(2))
      case s if s.startsWith("FZ") => Set("FZ", s.substring(2))
      case s if s.startsWith("BL") => Set("BL", s.substring(2))
      case s if s.startsWith("DR") => Set("DR", s.substring(2))
      case s if s.startsWith("BC") => Set("BC", s.substring(2))
      case s if s.startsWith("MI") => Set("MI", s.substring(2))
      case s if s.startsWith("PR") => Set("PR", s.substring(2))
      case s if s.startsWith("VC") => Set("VC", s.substring(2))
      case s if s.length == 4 => Set(s.substring(0, 2), s.substring(2, 4))
      case _ => Set.empty
    }
  }
  
  private def extractSimpleCodes(str: String): Set[String] = {
    if (str.length < 2) return Set.empty
    
    val codes = scala.collection.mutable.Set[String]()
    var i = 0
    while (i + 1 < str.length) {
      val code = str.substring(i, i + 2)
      if (code.matches("[A-Z]{2}")) {
        codes += code
      }
      i += 2
    }
    codes.toSet
  }
  
  def parseWeatherType(weatherType: String): WeatherCondition = {
    if (weatherType == null || weatherType.trim.isEmpty) {
      return WeatherCondition("NONE", 0, 0, 0, 0, 0, Set.empty)
    }
    
    val wt = weatherType.trim.toUpperCase
    
    val intensity = if (wt.startsWith("+")) "HEAVY"
                   else if (wt.startsWith("-")) "LIGHT"
                   else "MODERATE"
    
    val allCodes = extractWeatherCodes(wt)
    
    val hasThunderstorm = if (allCodes.contains("TS") || 
                             COMPOSITE_CODES.exists(c => c.startsWith("TS") && wt.contains(c))) 1 else 0
    
    val hasFreezingConditions = 
      if (allCodes.contains("FZ") || 
          FREEZING_PHENOMENA.exists(wt.contains) ||
          COMPOSITE_CODES.exists(c => c.startsWith("FZ") && wt.contains(c))) 1 else 0
    
    val hasPrecipitation = if (allCodes.exists(PRECIPITATION_CODES.contains)) 1 else 0
    
    val hasObscuration = if (allCodes.exists(OBSCURATION_CODES.contains)) 1 else 0
    
    val hasHazardousConditions = 
      if (allCodes.exists(HAZARDOUS_CODES.contains) || 
          hasThunderstorm == 1 || 
          hasFreezingConditions == 1) 1 else 0
    
    WeatherCondition(
      intensity = intensity,
      hasThunderstorm = hasThunderstorm,
      hasFreezingConditions = hasFreezingConditions,
      hasPrecipitation = hasPrecipitation,
      hasObscuration = hasObscuration,
      hasHazardousConditions = hasHazardousConditions,
      specificPhenomena = allCodes
    )
  }
  
  def hasFreezingPrecipitation(weatherType: String): Int = {
    if (weatherType == null || weatherType.isEmpty) return 0
    val wt = weatherType.toUpperCase
    if (FREEZING_PHENOMENA.exists(wt.contains)) 1 else 0
  }
  
  def hasVisibleMoisture(weatherType: String): Int = {
    if (weatherType == null || weatherType.isEmpty) return 0
    val codes = extractWeatherCodes(weatherType.toUpperCase)
    val moistureCodes = Set("RA", "DZ", "FG", "BR", "SH")
    if (codes.exists(moistureCodes.contains)) 1 else 0
  }
  
  def getWeatherHazardLevel(weatherType: String): Int = {
    if (weatherType == null || weatherType.isEmpty) return 0
    
    val parsed = parseWeatherType(weatherType)
    
    if (parsed.hasFreezingConditions == 1 && parsed.hasPrecipitation == 1) {
      3
    } else if (parsed.hasThunderstorm == 1) {
      3
    } else if (parsed.hasHazardousConditions == 1) {
      2
    } else if (parsed.intensity == "HEAVY" && parsed.hasPrecipitation == 1) {
      2
    } else if (parsed.hasPrecipitation == 1 || parsed.hasObscuration == 1) {
      1
    } else {
      0
    }
  }
  
  /**
   * VERSION OPTIMISÉE : Définir les UDFs en dehors de la méthode enrichWeatherTypeData
   * pour éviter la capture de l'objet WeatherTypeParser dans la closure
   */
  
  // Définir les fonctions UDF comme des val au niveau de l'objet
  @transient lazy val parseIntensityUDF = udf((wt: String) => {
    if (wt == null || wt.isEmpty) "NONE"
    else if (wt.startsWith("+")) "HEAVY"
    else if (wt.startsWith("-")) "LIGHT"
    else "MODERATE"
  })
  
  @transient lazy val hasThunderstormUDF = udf((wt: String) => {
    if (wt == null) 0
    else {
      val codes = extractWeatherCodes(wt)
      if (codes.contains("TS") || COMPOSITE_CODES.exists(c => c.startsWith("TS") && wt.toUpperCase.contains(c))) 1
      else 0
    }
  })
  
  @transient lazy val hasFreezingUDF = udf((wt: String) => {
    if (wt == null || wt.isEmpty) 0
    else {
      val wtUpper = wt.toUpperCase
      val codes = extractWeatherCodes(wtUpper)
      if (codes.contains("FZ") || 
          FREEZING_PHENOMENA.exists(wtUpper.contains) ||
          COMPOSITE_CODES.exists(c => c.startsWith("FZ") && wtUpper.contains(c))) 1
      else 0
    }
  })
  
  @transient lazy val hasFreezingPrecipUDF = udf((wt: String) => hasFreezingPrecipitation(wt))
  
  @transient lazy val hasVisibleMoistureUDF = udf((wt: String) => hasVisibleMoisture(wt))
  
  @transient lazy val hasPrecipitationUDF = udf((wt: String) => {
    if (wt == null || wt.isEmpty) 0
    else {
      val codes = extractWeatherCodes(wt.toUpperCase)
      if (codes.exists(PRECIPITATION_CODES.contains)) 1 else 0
    }
  })
  
  @transient lazy val hasObscurationUDF = udf((wt: String) => {
    if (wt == null || wt.isEmpty) 0
    else {
      val codes = extractWeatherCodes(wt.toUpperCase)
      if (codes.exists(OBSCURATION_CODES.contains)) 1 else 0
    }
  })
  
  @transient lazy val hasHazardousUDF = udf((wt: String) => {
    if (wt == null || wt.isEmpty) 0
    else {
      val codes = extractWeatherCodes(wt.toUpperCase)
      val wtUpper = wt.toUpperCase
      val hasHazard = codes.exists(HAZARDOUS_CODES.contains)
      val hasTS = codes.contains("TS") || COMPOSITE_CODES.exists(c => c.startsWith("TS") && wtUpper.contains(c))
      val hasFZ = codes.contains("FZ") || FREEZING_PHENOMENA.exists(wtUpper.contains)
      if (hasHazard || hasTS || hasFZ) 1 else 0
    }
  })
  
  @transient lazy val weatherHazardLevelUDF = udf((wt: String) => getWeatherHazardLevel(wt))
  
  @transient lazy val extractedCodesUDF = udf((wt: String) => {
    if (wt == null || wt.isEmpty) ""
    else extractWeatherCodes(wt.toUpperCase).mkString(",")
  })
  
  /**
   * Enrichit le DataFrame avec toutes les features météo
   */
  def enrichWeatherTypeData(weatherDF: DataFrame): DataFrame = {
    weatherDF
      .withColumn("weather_intensity", parseIntensityUDF(col("WeatherType")))
      .withColumn("has_thunderstorm", hasThunderstormUDF(col("WeatherType")).cast(IntegerType))
      .withColumn("has_freezing", hasFreezingUDF(col("WeatherType")).cast(IntegerType))
      .withColumn("has_freezing_precip", hasFreezingPrecipUDF(col("WeatherType")).cast(IntegerType))
      .withColumn("has_visible_moisture", hasVisibleMoistureUDF(col("WeatherType")).cast(IntegerType))
      .withColumn("has_precipitation", hasPrecipitationUDF(col("WeatherType")).cast(IntegerType))
      .withColumn("has_obscuration", hasObscurationUDF(col("WeatherType")).cast(IntegerType))
      .withColumn("has_hazardous", hasHazardousUDF(col("WeatherType")).cast(IntegerType))
      .withColumn("weather_hazard_level", weatherHazardLevelUDF(col("WeatherType")).cast(IntegerType))
      .withColumn("extracted_codes", extractedCodesUDF(col("WeatherType")))
  }
}

// ============================================
// INTÉGRATION ICING RISK
// ============================================

object IcingRiskFeatureV2 extends Serializable {
  
  def addIcingRiskFeatures(weatherDF: DataFrame): DataFrame = {
    
    import WeatherTypeParser._
    
    val enrichedDF = enrichWeatherTypeData(weatherDF)
    
    val icingRiskFlag = when(
      col("has_freezing_precip") === 1, 1
    ).when(
      col("DryBulbCelsius").isNotNull &&
      col("DryBulbCelsius").between(-15.0, 2.0) &&
      (
        (col("RelativeHumidity").isNotNull && col("RelativeHumidity") > 80.0) ||
        col("has_visible_moisture") === 1
      ), 1
    ).otherwise(0)
    
    val icingRiskLevel = when(
      col("has_freezing_precip") === 1, 3
    ).when(
      col("DryBulbCelsius").between(-10.0, 0.0) &&
      col("has_visible_moisture") === 1 &&
      col("RelativeHumidity") > 85.0, 2
    ).when(
      col("DryBulbCelsius").between(-15.0, 2.0) &&
      (col("RelativeHumidity") > 80.0 || col("has_visible_moisture") === 1), 1
    ).otherwise(0)
    
    enrichedDF
      .withColumn("Icing_Risk_Flag", icingRiskFlag.cast(IntegerType))
      .withColumn("Icing_Risk_Level", icingRiskLevel.cast(IntegerType))
  }
}

// ============================================
// TEST UNITAIRE
// ============================================

object WeatherTypeParserTest {
  
  def main(args: Array[String]): Unit = {
    
    val testCases = Seq(
      "+TS RAGR FG SQ",
      "+TSRA BR",
      "-FZDZ BR",
      "+SN FZFG",
      "SHRA BR VCTS",
      "+RA BR BCFG",
      "BLSN",
      "+FC -RA",
      "RAGR",
      "TSRA",
      "+TS GRRA BR SQ",
      "-FZRA -SN BR",
      "+DZ BR",
      ""
    )
    
    println("=" * 80)
    println("TEST DE PARSING DES CODES MÉTÉO")
    println("=" * 80)
    println()
    
    testCases.foreach { wt =>
      val parsed = WeatherTypeParser.parseWeatherType(wt)
      val codes = WeatherTypeParser.extractWeatherCodes(wt)
      
      println(s"Input: '$wt'")
      println(s"  ├─ Codes extraits: ${if (codes.isEmpty) "AUCUN" else codes.mkString(", ")}")
      println(s"  ├─ Intensité: ${parsed.intensity}")
      println(s"  ├─ Orage (TS): ${parsed.hasThunderstorm}")
      println(s"  ├─ Givrage (FZ): ${parsed.hasFreezingConditions}")
      println(s"  ├─ Précipitation: ${parsed.hasPrecipitation}")
      println(s"  ├─ Obscuration: ${parsed.hasObscuration}")
      println(s"  └─ Dangereux: ${parsed.hasHazardousConditions}")
      println()
    }
  }
}

WeatherTypeParserTest.main(Array())
```

```scala
import org.apache.spark.sql.{DataFrame, SparkSession}
import org.apache.spark.sql.functions._
import org.apache.spark.sql.types._
import org.apache.spark.ml.feature.{VectorAssembler, StringIndexer}
import org.apache.spark.ml.Pipeline

object WeatherTypeFeatureGenerator extends Serializable {
  

  def createFeatures(weatherDF: DataFrame): DataFrame = {
    
    weatherDF
      // 1. INTENSITÉ (String Indexing pour ML)
      .withColumn("intensity_heavy", 
        when(col("WeatherType").startsWith("+"), 1).otherwise(0).cast(IntegerType))
      .withColumn("intensity_light", 
        when(col("WeatherType").startsWith("-"), 1).otherwise(0).cast(IntegerType))
      
      // 2. ORAGE (TS)
      .withColumn("has_thunderstorm",
        when(col("WeatherType").contains("TS"), 1).otherwise(0).cast(IntegerType))
      
      // 3. GIVRAGE CRITIQUE - Précipitations givrantes
      .withColumn("has_freezing_precip",
        when(col("WeatherType").contains("FZRA") ||
             col("WeatherType").contains("FZDZ") ||
             col("WeatherType").contains("FZFG") ||
             col("WeatherType").contains("FZBR"), 1)
        .otherwise(0).cast(IntegerType))
      
      // 4. GIVRAGE GÉNÉRAL (FZ)
      .withColumn("has_freezing",
        when(col("WeatherType").contains("FZ"), 1).otherwise(0).cast(IntegerType))
      
      // 5. PRÉCIPITATIONS
      .withColumn("has_precipitation",
        when(col("WeatherType").contains("RA") ||  // Rain
             col("WeatherType").contains("DZ") ||  // Drizzle
             col("WeatherType").contains("SN") ||  // Snow
             col("WeatherType").contains("GR") ||  // Hail
             col("WeatherType").contains("GS") ||  // Small Hail
             col("WeatherType").contains("PL") ||  // Ice Pellets
             col("WeatherType").contains("SG") ||  // Snow Grains
             col("WeatherType").contains("IC") ||  // Ice Crystals
             col("WeatherType").contains("UP"), 1) // Unknown Precip
        .otherwise(0).cast(IntegerType))
      
      // 6. OBSCURATION (Visibilité réduite)
      .withColumn("has_obscuration",
        when(col("WeatherType").contains("BR") ||  // Mist
             col("WeatherType").contains("FG") ||  // Fog
             col("WeatherType").contains("FU") ||  // Smoke
             col("WeatherType").contains("HZ") ||  // Haze
             col("WeatherType").contains("DU") ||  // Dust
             col("WeatherType").contains("SA") ||  // Sand
             col("WeatherType").contains("VA"), 1) // Volcanic Ash
        .otherwise(0).cast(IntegerType))
      
      // 7. HUMIDITÉ VISIBLE (pour Icing Risk)
      .withColumn("has_visible_moisture",
        when(col("WeatherType").contains("RA") ||
             col("WeatherType").contains("DZ") ||
             col("WeatherType").contains("FG") ||
             col("WeatherType").contains("BR") ||
             col("WeatherType").contains("SH"), 1)
        .otherwise(0).cast(IntegerType))
      
      // 8. CONDITIONS DANGEREUSES
      .withColumn("has_hazardous",
        when(col("WeatherType").contains("DS") ||  // Dust Storm
             col("WeatherType").contains("SS") ||  // Sand Storm
             col("WeatherType").contains("FC") ||  // Funnel Cloud
             col("WeatherType").contains("SQ") ||  // Squall
             col("WeatherType").contains("PO") ||  // Dust Whirls
             col("WeatherType").contains("TS") ||  // Thunderstorm
             col("WeatherType").contains("FZ"), 1) // Freezing
        .otherwise(0).cast(IntegerType))
      
      // 9. TYPES SPÉCIFIQUES DE PRÉCIPITATIONS
      .withColumn("has_rain",
        when(col("WeatherType").contains("RA"), 1).otherwise(0).cast(IntegerType))
      .withColumn("has_snow",
        when(col("WeatherType").contains("SN"), 1).otherwise(0).cast(IntegerType))
      .withColumn("has_hail",
        when(col("WeatherType").contains("GR") || col("WeatherType").contains("GS"), 1)
        .otherwise(0).cast(IntegerType))
      
      // 10. NIVEAU DE DANGER MÉTÉO (0-3)
      .withColumn("weather_hazard_level",
        when(col("has_freezing_precip") === 1, 3)
        .when(col("has_thunderstorm") === 1, 3)
        .when(col("has_hazardous") === 1, 2)
        .when(col("intensity_heavy") === 1 && col("has_precipitation") === 1, 2)
        .when(col("has_precipitation") === 1 || col("has_obscuration") === 1, 1)
        .otherwise(0).cast(IntegerType))
  }
  
  def addIcingRiskFeatures(weatherDF: DataFrame): DataFrame = {
    
    // D'abord créer les features de base
    val withFeatures = createWeatherFeatures(weatherDF)
    
    withFeatures
      // ICING RISK FLAG (binaire 0/1)
      .withColumn("Icing_Risk_Flag",
        when(col("has_freezing_precip") === 1, 1)
        .when(col("DryBulbCelsius").isNotNull &&
              col("DryBulbCelsius").between(-15.0, 2.0) &&
              ((col("RelativeHumidity").isNotNull && col("RelativeHumidity") > 80.0) ||
               col("has_visible_moisture") === 1), 1)
        .otherwise(0).cast(IntegerType))
      
      // ICING RISK LEVEL (0-3)
      .withColumn("Icing_Risk_Level",
        when(col("has_freezing_precip") === 1, 3)
        .when(col("DryBulbCelsius").between(-10.0, 0.0) &&
              col("has_visible_moisture") === 1 &&
              col("RelativeHumidity") > 85.0, 2)
        .when(col("DryBulbCelsius").between(-15.0, 2.0) &&
              (col("RelativeHumidity") > 80.0 || col("has_visible_moisture") === 1), 1)
        .otherwise(0).cast(IntegerType))
  }
}
```
