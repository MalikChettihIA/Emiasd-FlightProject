## Flight & Weather Preprocesses Join

```scala
import com.flightdelay.features.FeaturePipeline
import com.flightdelay.features.balancer.DelayBalancedDatasetBuilder

val labeledFlightData =  DelayBalancedDatasetBuilder.prepareLabeledDataset(
  df = flightData,
  dxCol = experiment.featureExtraction.dxCol,
  delayThresholdMin = experiment.featureExtraction.delayThresholdMin,
  filterOnDxEquals1 = false  // Keep both delayed and on-time flights
)

val joinedData = FeaturePipeline.join(labeledFlightData, weatherData, experiment)
```

```scala
println("Flight Data: ", flightData.count())
println("Flight With Labels Data: ", labeledFlightData.count())
println("Joined Data: ", joinedData.count())
```

```scala
import com.flightdelay.features.joiners.FlightWeatherDataJoiner

val flightFeaturesWithTarget = experiment.featureExtraction.flightSelectedFeatures.map { features =>
      val featureNames = features.keys.toSeq
      if (featureNames.contains(experiment.target)) {
        featureNames
      } else {
        println(s"  - Automatically adding target '${experiment.target}' to flight features")
        featureNames :+ experiment.target
      }
    }

val joinedData2 = FlightWeatherDataJoiner.joinFlightsWithWeather(
  labeledFlightData,
  weatherData,
  weatherOriginDepthHours= experiment.featureExtraction.weatherOriginDepthHours,
  weatherDestinationDepthHours= experiment.featureExtraction.weatherDestinationDepthHours,
  removeLeakageColumns = true,
  flightFeaturesWithTarget,
  experiment.featureExtraction.weatherSelectedFeatures.map(_.keys.toSeq))
```

```scala
println("Joined Data 2: ", joinedData2.count())
```

```scala
import com.flightdelay.features.joiners.FlightWeatherDataJoiner

val joinedData3 = FlightWeatherDataJoiner.joinWeatherForAirport(
    labeledFlightData,
    weatherData,
    "origin",
    experiment.featureExtraction.weatherOriginDepthHours,
    flightFeaturesWithTarget,
    experiment.featureExtraction.weatherSelectedFeatures.map(_.keys.toSeq)
  )

println("Joined Data 3: ", joinedData3.count())
```

```scala
import org.apache.spark.sql.DataFrame
import org.apache.spark.sql.functions._
import org.apache.spark.sql.types._

// === 1) Harmonisation des clés ===

// flights: ORIGIN_WBAN: string, UTC_FL_DATE: date, feature_utc_departure_hour_rounded: long
val flights = labeledFlightData
  .withColumn("JOIN_WBAN", col("ORIGIN_WBAN"))
  .withColumn("JOIN_DATE", col("UTC_FL_DATE"))
  // on formate l'heure en "HHmm" (4 chiffres) pour matcher weather.Time (string "0000","0100",...)
  .withColumn(
    "JOIN_HOUR",
    lpad(col("feature_utc_departure_hour_rounded").cast(StringType), 4, "0")
  )

// weather: WBAN: string, Date: date, Time: string (format "HHmm")
val weatherKeys = weatherData
  .withColumn("W_WBAN", col("WBAN"))
  .withColumn("W_DATE", col("Date"))
  .withColumn("W_TIME", lpad(col("Time"), 4, "0"))
  .select("W_WBAN","W_DATE","W_TIME")

// === 2) Référentiels météo ===
val refWBAN         = weatherKeys.select("W_WBAN").distinct()                                 // station connue
val refWBANDate     = weatherKeys.select("W_WBAN","W_DATE").distinct()                        // station + date présente
val refWBANDateHour = weatherKeys.select("W_WBAN","W_DATE","W_TIME").distinct()               // station + date + heure

// === 3) Marquage des raisons d'absence ===
val withStation = flights
  .join(broadcast(refWBAN).withColumn("has_station", lit(true)),
        col("JOIN_WBAN") === col("W_WBAN"), "left")
  .drop("W_WBAN")

val withDate = withStation
  .join(broadcast(refWBANDate).withColumn("has_date", lit(true)),
        col("JOIN_WBAN") === col("W_WBAN") && col("JOIN_DATE") === col("W_DATE"),
        "left")
  .drop("W_WBAN","W_DATE")

val withHour = withDate
  .join(broadcast(refWBANDateHour).withColumn("has_hour", lit(true)),
        col("JOIN_WBAN") === col("W_WBAN") &&
        col("JOIN_DATE") === col("W_DATE") &&
        col("JOIN_HOUR") === col("W_TIME"),
        "left")
  .drop("W_WBAN","W_DATE","W_TIME")

val diagnosed = withHour
  .withColumn("has_station", coalesce(col("has_station"), lit(false)))
  .withColumn("has_date",    coalesce(col("has_date"),    lit(false)))
  .withColumn("has_hour",    coalesce(col("has_hour"),    lit(false)))
  .withColumn(
    "join_reason",
    when(!col("has_station"), lit("missing_station")) // pas de météo pour cet aéroport
    .when(col("has_station") && !col("has_date"), lit("missing_date")) // station ok, date absente
    .when(col("has_station") && col("has_date") && !col("has_hour"), lit("missing_hour")) // heure absente
    .otherwise(lit("matched"))
  )

// === 4) Comptes & récapitulatif ===
val totalFlights     = diagnosed.count()
val matchedCount     = diagnosed.filter(col("join_reason") === "matched").count()
val missingStationCt = diagnosed.filter(col("join_reason") === "missing_station").count()
val missingDateCt    = diagnosed.filter(col("join_reason") === "missing_date").count()
val missingHourCt    = diagnosed.filter(col("join_reason") === "missing_hour").count()

def pct(n: Long) = f"${n * 100.0 / totalFlights}%.3f%%"

println("\n===== Join Coverage Report (Flights → Weather) =====")
println(f"Total flights:              $totalFlights%,d")
println(f"Matched (full key):         $matchedCount%,d   (${pct(matchedCount)})")
println(f"Missing station (WBAN):     $missingStationCt%,d   (${pct(missingStationCt)})")
println(f"Missing date (WBAN+Date):   $missingDateCt%,d      (${pct(missingDateCt)})")
println(f"Missing hour (WBAN+Date+H): $missingHourCt%,d      (${pct(missingHourCt)})")

// Top WBAN manquantes (pas de station dans weather)
println("\nTop WBAN with missing_station:")
diagnosed.filter(col("join_reason")==="missing_station")
  .groupBy("JOIN_WBAN").count().orderBy(desc("count")).show(10, truncate=false)

// Exemples de dates manquantes pour stations connues
println("\nExamples where station exists but date is missing:")
diagnosed.filter(col("join_reason")==="missing_date")
  .select("JOIN_WBAN","JOIN_DATE").distinct().orderBy(desc("JOIN_DATE")).show(10, truncate=false)

// Exemples d’heures manquantes pour WBAN+Date présents
println("\nExamples where WBAN+Date exists but hour is missing:")
diagnosed.filter(col("join_reason")==="missing_hour")
  .select("JOIN_WBAN","JOIN_DATE","JOIN_HOUR").show(10, truncate=false)
```
