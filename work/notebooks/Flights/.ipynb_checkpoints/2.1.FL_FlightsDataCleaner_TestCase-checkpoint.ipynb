```scala
import org.apache.spark.sql.DataFrame
import org.apache.spark.sql.functions._

def explainArrDelayNewNulls(df: DataFrame): Unit = {
  // 0) Taille & nulls globaux
  val total = df.count()
  println(s"Total rows: $total")

  df.select(
      sum(when(col("ARR_DELAY_NEW").isNull, 1).otherwise(0)).as("nulls"),
      sum(when(col("ARR_DELAY_NEW").isNotNull, 1).otherwise(0)).as("non_nulls")
    )
    .withColumn("null_rate", col("nulls")/lit(total))
    .show(false)

  // 1) Hypothèse standard (dataset US BTS) :
  //    ARR_DELAY_NEW est NULL si le vol est annulé (CANCELLED=1) ou détourné (DIVERTED=1)
  val expectedNull = col("CANCELLED") === lit(1.0) || col("DIVERTED") === lit(1.0)

  df.select(
      sum(when(expectedNull && col("ARR_DELAY_NEW").isNull, 1).otherwise(0)).as("null_expected"),
      sum(when(!expectedNull && col("ARR_DELAY_NEW").isNull, 1).otherwise(0)).as("null_unexpected"),
      sum(when(expectedNull && col("ARR_DELAY_NEW").isNotNull, 1).otherwise(0)).as("notnull_unexpected")
    )
    .withColumn("null_expected_rate", col("null_expected")/lit(total))
    .withColumn("null_unexpected_rate", col("null_unexpected")/lit(total))
    .withColumn("notnull_unexpected_rate", col("notnull_unexpected")/lit(total))
    .show(false)

  // 2) Répartition des NULLs par statut annulé/détourné
  df.groupBy(col("CANCELLED"), col("DIVERTED"))
    .agg(
      count(lit(1)).as("rows"),
      sum(when(col("ARR_DELAY_NEW").isNull, 1).otherwise(0)).as("nulls")
    )
    .withColumn("null_rate", col("nulls")/col("rows"))
    .orderBy(desc("null_rate"))
    .show(false)

  // 3) Cas problématiques à inspecter
  // 3a) NULL mais vol ni annulé ni détourné (unexpected)
  val unexpectedNull = df
    .filter(!expectedNull && col("ARR_DELAY_NEW").isNull)
    .select("FL_DATE","OP_CARRIER","FL_NUM","ORIGIN","DEST",
            "CANCELLED","DIVERTED","ARR_TIME","ARR_DELAY","ARR_DELAY_NEW")
  println(s"Unexpected NULL rows (sample): ${unexpectedNull.count()}")
  unexpectedNull.show(20, truncate = false)

  // 3b) Non-NULL alors que annulé/détourné (incohérent)
  val inconsistentNotNull = df
    .filter(expectedNull && col("ARR_DELAY_NEW").isNotNull)
    .select("FL_DATE","OP_CARRIER","FL_NUM","ORIGIN","DEST",
            "CANCELLED","DIVERTED","ARR_TIME","ARR_DELAY","ARR_DELAY_NEW")
  println(s"Inconsistent NOT NULL on cancelled/diverted (sample): ${inconsistentNotNull.count()}")
  inconsistentNotNull.show(20, truncate = false)

  // 4) Autres pistes fréquentes : champs d’arrivée manquants
  df.select(
      sum(when(col("ARR_TIME").isNull, 1).otherwise(0)).as("arr_time_nulls"),
      sum(when(col("ARR_DELAY").isNull, 1).otherwise(0)).as("arr_delay_nulls"),
      sum(when(col("ARR_DELAY_NEW").isNull, 1).otherwise(0)).as("arr_delay_new_nulls")
    ).show(false)

  // Lien entre ARR_TIME manquant et ARR_DELAY_NEW manquant
  df.groupBy(col("ARR_TIME").isNull.as("arr_time_is_null"))
    .agg(
      count(lit(1)).as("rows"),
      sum(when(col("ARR_DELAY_NEW").isNull, 1).otherwise(0)).as("arr_delay_new_nulls")
    )
    .withColumn("null_rate", col("arr_delay_new_nulls")/col("rows"))
    .orderBy(desc("null_rate"))
    .show(false)

  // 5) Où ça se concentre ? (par compagnie, aéroport, date)
  df.filter(col("ARR_DELAY_NEW").isNull)
    .groupBy("OP_CARRIER")
    .count()
    .orderBy(desc("count"))
    .show(20, truncate = false)

  df.filter(col("ARR_DELAY_NEW").isNull)
    .groupBy("FL_DATE")
    .count()
    .orderBy(desc("count"))
    .show(20, truncate = false)
}

explainArrDelayNewNulls(flightData)
```
